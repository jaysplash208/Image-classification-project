```python
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("---step 1: Loading the Pre-Trained 'Brain'---")
model = VGG16(weights='imagenet')
print("VGG Model Loaded succesfully")
```

```python
print("\n--- Step 2: Preparing our custom image ---")
img_path = '/content/Adesua_Etomi.jpg'
img = image.load_img(img_path, target_size=(224,224))

x = image.img_to_array(img)

x = np.expand_dims(x, axis=0)

x = preprocess_input(x)

print("Image Converted To Mathematical Array and Preprocessed")
```

```python
print("\n--- AI interferance (Making Predictions) ---")

print(("Analyzing Visual Data......."))

predictions = model.predict(x)

result = decode_predictions(predictions, top =3)[0]
```

```python
print("\n--- Ai Classification Results ---")
for pred_result in result:
  label = pred_result[1]
  confidence = pred_result[2] * 100
  print(f"Prediciton: {label} (Confidence: {confidence:.2f}%")

  plt.imshow(image.load_img(img_path))
  plt.axis('off')
  plt.show()
```
